```python
import h5py

def check_edge_index_in_hdf5(path):
    path_to_hdf5_file, sample_id = os.path.split(path)
    with h5py.File(path_to_hdf5_file, 'r') as f:
        if sample_id not in f:
            print(f"❌ Sample ID '{sample_id}' not found in HDF5.")
            return
        
        g = f[sample_id]
        edge_keys = [k for k in g.keys() if 'edge' in k.lower()]
        if edge_keys:
            print(f"✅ Found potential edge-related keys: {edge_keys}")
            for k in edge_keys:
                print(f"    ➤ {k} shape = {g[k].shape}, dtype = {g[k].dtype}")
        else:
            print("❌ No edge-related keys found in this sample.")


check_edge_index_in_hdf5("/ehome/xinyi/Xinyi_GNN_aorta/data/vessel-dataset/single/raw/database.hdf5/sample_0000")
```
